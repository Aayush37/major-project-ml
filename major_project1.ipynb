```
 from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```
import pandas as pd
df = pd.read_table('/content/drive/MyDrive/Training/ML/Untitled folder/amazon_alexa.tsv')
reviews=df.verified_reviews
ratings=df.rating
```

```
df
```

```
def dec(b):
  if (b<=5 and b>=4):
    return('positive')
  if (b>=1 and b<=2):
    return('negative')
  if (b==3):
    return('neutral')
```

```
def decode(a):
  if (a>0):
    return('positive')
  if (a<0):
    return('negative')
  if (a==0):
    return('neutral')
```

```
analyzer = SentimentIntensityAnalyzer()
```

```

sentiment=[]
for review in reviews:
    vs = analyzer.polarity_scores(review)
    vs=str(vs)
    temp=vs.split()
    temp=temp[7].split('}')
    sen=decode(float(temp[0]))
    sentiment.append(str(sen))
sentiment
```

```
sen_test=[]
for rating in ratings:
  k=dec(int(rating))
  sen_test.append(k)
```

```
import numpy as np
sen_pred=sentiment
sen_test=np.array(sen_test)
sen_pred=np.array(sen_pred)
```

```
df['date'].unique()
```

```
from sklearn import metrics
from sklearn.metrics import confusion_matrix,classification_report
print("Accuracy : ", metrics.accuracy_score(sen_test,sen_pred))
confusion_matrix(sen_test,sen_pred)
```

```
print(classification_report(sen_test,sen_pred))
```
